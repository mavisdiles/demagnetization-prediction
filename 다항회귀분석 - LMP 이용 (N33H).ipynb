단순 다항회귀분석을 위해 종속변수를 1개로 고정

1. 형상은 하나로 (현재 코드에서는 2.03) 고정
2. 온도,시간을 LMP=(온도 +273)*(ln(시간) +120) 로 통일
3. 이렇게 결정된 LMP 변수를 바탕으로 다항식을 만들어 실험, 즉 y = (LMP)^n + (LMP)^(n-1) + ... 형태로 만듬
4. 2차 항부터 n차항까지 시각화

실험 결과,
 2차항에서 11차항으로 증가함에 따라 결정계수도 1에 가까워진다. (2차 항의 결정계수는 0.9565393525819005이고, 11차 항의 결정계수는 0.9684127812153065이다) 그러나 12차항부터는 다시 줄어든다(overfitting 으로 추측). 결굴, 11차항일때 가장 좋음

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plot
import numpy as np
import sklearn
import pandas as pd
import openpyxl
#plot.rcParams["font.family"] = 'Malgun gothic'

data = pd.read_excel("data/N33H.xlsx", header=1)
newData = data[['Pc', '온도', '시간', '감자율']].head(20) # pc = 2.03

#단순 회귀 모델 생성
model = LinearRegression()

#속성(변수)선택
temperature = newData['온도']
time = np.log(newData['시간'])
C = 120
newData['LMP'] = (temperature +273)*(time + C) # LMP=(온도 +273)*(ln(시간) + C)
x= newData[['LMP']]

y = newData[['감자율']]

def create_poly(hyper_degree):
    #변환
    poly = PolynomialFeatures(degree = hyper_degree) # n차함수 적용
    x_poly = poly.fit_transform(x) # 데이터를 n차식으로 변형

    model.fit(x_poly, y)

    #예측 모델 생성
    y_p = model.predict(x_poly)

    #출력
    print('변환 차수 : ', hyper_degree)
    #print(x_poly)
    #기울기와 y절편 출력
    print('기울기 : ', model.coef_)
    print('y절편 : ', model.intercept_)

    #결정계수
    relation_square = model.score(x_poly, y)
    print('결정계수 : ', relation_square)

    #그래프로 나타내기 
    plot.scatter(x, y, marker = '+')
    plot.scatter(x, y_p, marker = 'o')
    plot.show()
    
# 2차항 부터 20차항 까지 출력
for i in range(2,21):
    create_poly(i)
```
