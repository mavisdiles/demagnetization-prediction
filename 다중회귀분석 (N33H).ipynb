재질 N33H에 대해 분석

1. 감자율을 형상,온도,시간 각각의 독립변수를 하나와 비교했을 때 상관관계 크게 나타나지 않음.


2. 3개의 변수가 감자율에 대해 '선형적'이다 가정하고 다중회귀분석을 해본 결과 ( h(x1,x2,x3) = y)

- 회귀계수:  [0.0332772, -0.00405698, -0.00023226]
 => 즉, 형상,온도,시간의 기울기가 각각, 0.0332772,-0.00405698, -0.00023226이다.

- train, test를 8:2로 배분하여 훈련 후 R2 측정
=> train, test R2값이 모두 높게 나타나 예측률이 좋아 보인다.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')

df = pd.read_excel("data/N33H.xlsx", header=1)
x = df[['Pc', '온도', '시간']]
y = df[['감자율']]

def correlation():
    #상관관계 Pc
    plt.scatter(df[['Pc']], df[['감자율']], alpha=0.4)
    plt.xlabel("형상")
    plt.ylabel("감자율")
    plt.show()

    #상관관계 온도
    plt.scatter(df[['온도']], df[['감자율']], alpha=0.4)
    plt.xlabel("온도")
    plt.ylabel("감자율")
    plt.show()

    #상관관계 시간
    plt.scatter(df[['시간']], df[['감자율']], alpha=0.4)
    plt.xlabel("시간")
    plt.ylabel("감자율")
    plt.show()

def prediction(hyper_degree):
    # 데이터셋 분리
    x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, test_size=0.2) 
    
    #모델 생성
    model = LinearRegression()

    poly = PolynomialFeatures(degree = hyper_degree) # n차함수 적용
    x_train_poly = poly.fit_transform(x_train) # 데이터를 n차식으로 변형
    x_test_poly = poly.fit_transform(x_test) # 데이터를 n차식으로 변형
    model.fit(x_train_poly, y_train) # model fitting
    
    # K-fold model validation
    arr_score = cross_val_score(model,x_train_poly,y_train,scoring='r2',cv=10)
    
    
    #예측 비교
    y_predict = model.predict(x_test_poly)
    plt.scatter(y_test, y_predict, alpha=0.4)
    plt.xlabel("실제 감자율")
    plt.ylabel("예측 감자율")
    plt.title("MULTIPLE LINEAR REGRESSION")
    plt.show()
    
    print('변환 차수 : ', hyper_degree)
    print(f"cv10 R2:{arr_score}")
    print(f'cv10 평균점수 : {arr_score.mean()}')
    print("test set R2: ", model.score(x_test_poly, y_test))
    print("회귀계수: ", model.coef_)
    # print("train set R2: ", model.score(x_train_poly, y_train))
    

# 2차항 부터 10차항 까지 출력
for i in range(2,8):
    prediction(i)
#prediction(2)
```
